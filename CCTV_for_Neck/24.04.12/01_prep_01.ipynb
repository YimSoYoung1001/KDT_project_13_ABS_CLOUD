```python
max_pixel = 0
try:
    for i in range(1, 71):
        img = Image.open(f'./neck/negative/normal ({i}).jpg')
        img_size = img.size
        if img_size[0]*img_size[1] > max_pixel : 
            max_pixel = img_size[0]*img_size[1]
            print(img_size)
            print(max_pixel)
            print(f"img_file_name : normal ({i}).jpg ")
except:
    pass
```
